```python
# series 1D array
```

```python
import pandas as pd
```

```python
data = [100.1, 102.2, 104.3]

series = pd.Series(data)
print(series)
```

```python
data = ["a","b","c"]

series = pd.Series(data)
print(series)
```

```python
data = [True, False, True]

series = pd.Series(data)
print(series)
```

```python
data = [100, 102, 104]

series = pd.Series(data, index= [1,2,3])
print(series)
```

```python
data = [100, 102, 104]

series = pd.Series(data, index= ["a","b","c"])
print(series)
```

```python
data = [100, 102, 104]

series = pd.Series(data, index= ["apartment1","apartment2","apartment3"])

# loc location by label
print(series.loc["apartment1"])
```

```python
data = [100, 102, 104]

series = pd.Series(data, index= ["a","b","c"])

series.loc["c"] = 200 
print(series)
```

```python
data = [100, 102, 104]

series = pd.Series(data, index= ["a","b","c"])

# iloc integer location by label
print(series.iloc[0])
```

```python
data = [100.1, 102.2, 104.3, 205.6, 306.3]

series = pd.Series(data, index= ["a","b","c","d","e"])

print(series[series >= 200])
print()
print(series[series < 200])
```

```python
calories = {"Day 1" : 1750, "Day 2": 2100, "Day 3": 1700}

series = pd.Series(calories)

# cheated on my diet
series.loc["Day 3"] += 500

print(series.loc["Day 3"])
print()
print(series)
```

```python
calories = {"Day 1" : 1750, "Day 2": 2100, "Day 3": 1700}

series = pd.Series(calories)

print(series[series >= 2000])
print()
print()
print(series[series < 2000])
```
