```python
import pandas as pd
```

```python
data = {"Name" : ["Spongebob", "Patrick", "Squidward"], 
        "Age" : [30, 35, 50]
       }

df = pd.DataFrame(data, index = ["Emp1","Emp2","Emp3"])
print(df.loc["Emp3"])
print()
print(df.iloc[0]) # int location by label
print()
print(df)
```

```python
# Add a new column 

df["Job"] = ["Cook","N/A","Cashier"]
df
```

```python
# new row
# another df name new_row
new_row = pd.DataFrame([{"Name":  "Sandy","Age" : 28,"Job": "Engineer"},
                       {"Name":  "Eugene","Age" : 60,"Job": "Manager"}],
                      index =["Emp4", "Emp5"]) 
df = pd.concat([df, new_row])

print(df)
```
