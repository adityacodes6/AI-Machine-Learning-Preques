```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# relationship between two varibles
# helps identify correlation
# eg. study hours vs sleep score
```

```python
x = [0, 1, 1, 2, 3, 4, 5, 6, 7, 7, 8] # hours studied
y = [55, 60, 65, 62, 68, 70, 75, 78, 82, 85, 87] # grades

plt.scatter(x, y)
plt.xlabel("Hours Studied")
plt.ylabel("Grades")
plt.title("Test Scores")

plt.show()
```

```python
plt.scatter(x, y, color="skyblue",
                   alpha= 0.6) # transperancy


plt.xlabel("Hours Studied")
plt.ylabel("Grades")
plt.title("Test Scores")

plt.show()
```

```python
plt.scatter(x, y, color="skyblue",
                   alpha= 0.6,
                   s= 100) # s for size


plt.xlabel("Hours Studied")
plt.ylabel("Grades")
plt.title("Test Scores")

plt.show()
```

```python
# class 1
x = [0, 1, 1, 2, 3, 4, 5, 6, 7, 7, 8] # hours studied
y = [55, 60, 65, 62, 68, 70, 75, 78, 82, 85, 87] # grades

# class 2
x2 = [0, 1, 2, 2, 3, 4, 5, 6, 7, 8, 8] # hours studied
y2 = [50, 58, 65, 70, 72, 78, 83, 88, 92, 95, 97] # grades

```

```python
plt.scatter(x, y, color="green",
                   alpha= 0.6,
                   s= 100,
                   label = "Class A") # s for size


plt.scatter(x2, y2, color="skyblue",
                   alpha= 0.6,
                   s= 100,
                   label = "Class B") # s for size


plt.xlabel("Hours Studied")
plt.ylabel("Grades")
plt.title("Test Scores")
plt.legend()

plt.show()
```
