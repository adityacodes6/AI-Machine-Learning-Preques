```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# circular chart divided into slices to show percentages of the total
# good for visualizing distribution among categories
```

```python
categories = ["Freshman", "Sophomores", "Juniors", "Seniors"]
values = np.array([300, 250, 275, 225])

plt.pie(values)
plt.show()
```

```python
plt.pie(values, labels= categories)

plt.show()
```

```python
plt.pie(values, labels= categories,
                autopct= "%1f") # auto percentage

plt.show()
```

```python
plt.pie(values, labels= categories,
                autopct= "%1.1f") # to display 1 digit after decimal

plt.show()
```

```python
# ["Freshman", "Sophomores", "Juniors", "Seniors"]
colors = ["red", "yellow", "blue", "green"]

plt.pie(values, labels= categories,
                autopct= "%1.1f",
                colors = colors)

plt.show()
```

```python
# ["Freshman", "Sophomores", "Juniors", "Seniors"]
colors = ["red", "yellow", "blue", "green"]

plt.pie(values, labels= categories,
                autopct= "%1.1f",
                colors = colors)
plt.title("My College")

plt.show()
```

```python

plt.pie(values, labels= categories,
                autopct= "%1.1f",
                colors = colors,
                # explode means to display it slightly (highlight it)
                explode = [0, 0, 0, 0.1]) 

plt.title("My College")

plt.show()
```

```python

plt.pie(values, labels= categories,
                autopct= "%1.1f",
                colors = colors,
                # explode means to display it slightly (highlight it)
                explode = [0, 0, 0, 0.1],
                shadow= True) 

plt.title("My College")

plt.show()
```

```python

plt.pie(values, labels= categories,
                autopct= "%1.1f",
                colors = colors,
                # explode means to display it slightly (highlight it)
                explode = [0, 0, 0, 0.1],
                shadow= True,
                startangle = 90) 

plt.title("My College")

plt.show()
```
