```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([2023, 2024, 2025, 2026])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([17, 23, 38, 5])
y3 = np.array([12, 15, 20, 30])

# labels
plt.title("class size", fontsize= 26,
         family= "Arial",
         fontweight= "bold",
         color="darkblue")

plt.xlabel("year",  fontsize= 15,
          fontweight= "bold",
          color= "blue")

plt.ylabel("students",  fontsize= 15,
          fontweight= "bold",
          color= "blue")
plt.xticks(x) # remove float points

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

plt.show()
```

```python
x = np.array([2023, 2024, 2025, 2026])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([17, 23, 38, 5])
y3 = np.array([12, 15, 20, 30])

# labels
plt.title("class size", fontsize= 26,
         family= "Arial",
         fontweight= "bold",
         color="darkblue")

plt.xlabel("year",  fontsize= 15,
          fontweight= "bold",
          color= "blue")

plt.ylabel("students",  fontsize= 15,
          fontweight= "bold",
          color= "blue")

plt.tick_params(axis="both",
                 colors="red")

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

plt.show()
```
