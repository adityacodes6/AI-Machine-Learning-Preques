```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([2023, 2024, 2025, 2026])
y = np.array([15, 25, 30, 20])

#     x-axis, y-axis
plt.plot(x, y, marker= ".",
        markersize= 10,
        markerfacecolor= "green",
        markeredgecolor= "red",
        linestyle="solid",
        linewidth = 3,
        color= "darkblue")
plt.show()
```

```python
x1 = np.array([2023, 2024, 2025, 2026])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([17, 23, 38, 5])
#     x-axis, y-axis
plt.plot(x, y1, marker= ".",
        markersize= 10,
        markerfacecolor= "green",
        markeredgecolor= "red",
        linestyle="solid",
        linewidth = 3,
        color= "darkblue")

plt.plot(x, y2, marker= ".",
        markersize= 10,
        markerfacecolor= "green",
        markeredgecolor= "red",
        linestyle="solid",
        linewidth = 3,
        color= "darkblue")

plt.show()
```

```python
x1 = np.array([2023, 2024, 2025, 2026])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([17, 23, 38, 5])


line_style = dict(marker= ".",
                  markersize= 10,
                  markerfacecolor= "green",
                  markeredgecolor= "red",
                  linestyle="solid",
                  linewidth = 3,
                  color= "darkblue")

plt.plot(x, y1, **line_style)
plt.plot(x, y2, **line_style) # unpacking

plt.show()
```

```python
x1 = np.array([2023, 2024, 2025, 2026])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([17, 23, 38, 5])
y3 = np.array([12, 15, 20, 30])

line_style = dict(marker= ".",
                  markersize= 10,
                  markerfacecolor= "green",
                  markeredgecolor= "red",
                  linestyle="solid",
                  linewidth = 3,
                  color= "darkblue")

plt.plot(x, y1, **line_style)
plt.plot(x, y2, **line_style) # unpacking
plt.plot(x, y3, **line_style)

plt.show()
```

```python
x1 = np.array([2023, 2024, 2025, 2026])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([17, 23, 38, 5])
y3 = np.array([12, 15, 20, 30])

line_style = dict(marker= ".",
                  markersize= 10,
                  markerfacecolor= "green",
                  markeredgecolor= "red",
                  linestyle="solid",
                  linewidth = 3)

plt.plot(x, y1, color="red", **line_style)
plt.plot(x, y2, color= "green", **line_style) # unpacking
plt.plot(x, y3, color= "blue", **line_style)

plt.show()
```
