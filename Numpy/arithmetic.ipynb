```python
arr = ([1.01,2.5,3.99])
print(np.sqrt(arr))
print(np.round(arr)) # nearest int
print(np.floor(arr)) # down
print(np.ceil(arr)) # up
```

```python
print(np.pi)
```

```python
#exercise

radius = np.array([1,2,3])
print(np.pi * radius ** 2)
```

```python
# element wise arithmetic

arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])

print(arr1 + arr2)
print(arr1 - arr2)
print(arr1 * arr2)
print(arr1 / arr2)
print(arr1 ** arr2)
```

```python
# comparison opr

scores = np.array([91,55,100,73,82,64])
print(scores == 100)
print(scores >= 60) # 5 students pass
print(scores < 60) # 1 student fail
```

```python
scores = np.array([91,55,100,73,82,64])

# subscript opr
scores[scores < 60] = 0
print(scores)
```
