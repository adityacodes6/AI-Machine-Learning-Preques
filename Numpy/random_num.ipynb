```python
import numpy as np
```

```python
rng = np.random.default_rng(seed = 1)

print(rng.integers(1, 7)) # random no between 1 to 6
print(rng.integers(low= 1, high= 101, size = (3, 2)))
```

```python
print(np.random.uniform(low = -1, high=1, size =3))
# uniform dist each value has equal chance of being selected
```

```python
np.random(seed=1)
print(np.random.uniform(low = -1, high=1, size =(3, 2)))
```

```python
# shuffle
```

```python
arr = np.array([1,2,3,4,5,6])
print(arr)
print()
rng = np.random.default_rng()
rng.shuffle(arr)
print(arr)
```

```python
fruits = np.array(["apple", "banana", "orange", "coconut", "pineapple", "kiwi"])
fruits = rng.choice(fruits, size= (3, 3))
print(fruits)
```

```python
a = np.arange(1, 10)
print(a)
```

```python
# evenly spaced values
l = np.linspace(1, 10, 2)
print(l)
```
