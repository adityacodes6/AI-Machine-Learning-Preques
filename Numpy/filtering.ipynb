```python
import numpy as np
```

```python
# filtering refers to process of selecting elements from an array that match a given condition
```

```python
ages = np.array([[21, 17, 19, 20, 16, 30, 18, 65],
                 [39, 22, 15, 99, 18, 19, 20, 21]])

teenagers = ages[ages < 18] #bool indexing
print(teenagers)
```

```python
print(ages) # original not affected
```

```python
adults = ages[(ages >= 18) & (ages < 65)] 
print(adults)
```

```python
seniors = ages[ages >= 65]
print(seniors)
```

```python
evens = ages[ages % 2 == 0]
print(evens)
```

```python
odds = ages[ages % 2 != 0]
print(odds)
```

```python
# where funtion to filter

adults = np.where(ages >= 18, ages, 0)
print(adults)
```
